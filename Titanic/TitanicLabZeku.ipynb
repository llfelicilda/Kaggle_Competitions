```python
from zeku import feature_engineering, pmml, metrics, data_source, utils
from sklearn import linear_model, tree, ensemble, decomposition, datasets
from pandas import DataFrame, MultiIndex
import pandas as pd
import numpy as np
from scipy.stats import mode

pd.set_option('max_column', 100)
```

```python
data_1 = titanic_data.data[['name', 'sex', 'pclass', 'age', 'sibsp', 'parch', 'fare', 'survived','embarked']]
```

```python
data_1['embarked'] = data_1['embarked'].fillna('S')
utils.multiapply(data_1, {'pclass': str}, True)
data_dict_list = utils.input_data_schema(data_1, ['name'], 'survived')
```

```python
aggregation_schema = {'famsize':{'columns':['parch','sibsp'],'function':np.sum,'function name':'sum'}}
agg_df, agg_trans = feature_engineering.create_aggregations(data_1,aggregation_schema)
```

```python
regex_schema = {'name': ['.*Miss\..*', '.*Mr\..*', '.*Mrs\..*', '.*Master\..*']}
regex_frame, regex_transformations = feature_engineering.match_from_schema(agg_df, regex_schema, False, transformations = agg_trans)
regex_frame = regex_frame.drop(['name'],axis=1)
regex_frame.head()
```

```python
regex_frame
```

### The problem occurs in these two cells. While in our pandas dataframes, the 'regex' columns appear as 0's and 1's, in the PMML model, they have to be recorded as booleans ('true' or 'false'). So when we create the age_mapping below, we are telling the PMML model to use 0's and 1's, when it is expecting to find 'false' and 'true', respectively. We can get around it though by using some clever Python.

```python
cols = list(regex_frame.columns)

fare_mapping = regex_frame.pivot_table(rows=['embarked', 'pclass'], values='fare', aggfunc=np.median)
age_mapping = regex_frame.pivot_table(rows=['sex', 'name: .*Miss\..*','name: .*Mr\..*','name: .*Mrs\..*','name: .*Master\..*']
                                      , values='age', aggfunc=np.mean)
age_mapping
```

```python
df_mapping, transf = feature_engineering.map_values(regex_frame, {'fare_fill': fare_mapping,
                                                                'age_fill': age_mapping},transformations = regex_transformations)
df_filled, fill_transf = feature_engineering.fill_missing(df_mapping, {'age_filled': {'source': 'age', 'fill': 'age_fill'},
                                                                       'fare_filled': {'source': 'fare', 'fill': 'fare_fill'}},
                                                          transformations=transf)
df_filled.head()
```

### We are first going to create a temporary frame called 'tmp_regex'. We are going to replace the regex column values (which are 0's and 1's) with string values ('false' or 'true'). Then when we make the pivot_table mappings, they will reference strings instead of ints.

```python
cols = list(regex_frame.columns)
tmp_regex = regex_frame[cols].copy()
tmp_regex[['name: .*Miss\..*','name: .*Mr\..*','name: .*Mrs\..*','name: .*Master\..*']] = tmp_regex[['name: .*Miss\..*','name: .*Mr\..*','name: .*Mrs\..*','name: .*Master\..*']].applymap(lambda x: str(bool(x)).lower())
fare_mapping = tmp_regex.pivot_table(rows=['embarked', 'pclass'], values='fare', aggfunc=np.median)
age_mapping = tmp_regex.pivot_table(rows=['sex', 'name: .*Miss\..*','name: .*Mr\..*','name: .*Mrs\..*','name: .*Master\..*']
                                      , values='age', aggfunc=np.mean)
age_mapping
```

### We now create the map_values transformation, along with the new columns ('fare_fill' and 'age_fill'). But now we need to transform the regex columns back to 0's and 1's. We do this by capitalizing the strings and then using 'eval' to get a boolean, and then wrapping that with 'int'. Now we have the same dataframe as before, but the transformations element for the age mapping will be valid for the PMML file. It's ugly, but it works.

```python
df_mapping, transf = feature_engineering.map_values(tmp_regex, {'fare_fill': fare_mapping,
                                                                'age_fill': age_mapping},transformations = regex_transformations)
tmp_df_mapping = df_mapping.copy()
tmp_df_mapping[['name: .*Miss\..*','name: .*Mr\..*','name: .*Mrs\..*','name: .*Master\..*']] = tmp_df_mapping[['name: .*Miss\..*','name: .*Mr\..*','name: .*Mrs\..*','name: .*Master\..*']].applymap(lambda x: int(eval(x.capitalize())))
df_filled, fill_transf = feature_engineering.fill_missing(tmp_df_mapping, {'age_filled': {'source': 'age', 'fill': 'age_fill'},
                                                                       'fare_filled': {'source': 'fare', 'fill': 'fare_fill'}},
                                                          transformations=transf)
df_filled.head()
```

```python
cols = list(df_filled.columns)
cols.remove('age')
cols.remove('age_fill')
cols.remove('fare')
cols.remove('fare_fill')
df_filled = df_filled.fillna(df_filled.mean())[cols]
df_filled.head()
```

```python
disc_schema = {'age_filled': [[0.0,12.0],[12.0,35.0],[35.0,np.inf]]
               ,'fare_filled': [[0.0,15.0],[15.0,35.0],[35.0,100.0],[100.0,np.inf]]}
print disc_schema
discretized_frame, ef_transformations = feature_engineering.discretize_from_schema(df_filled[cols],
                                                                                   disc_schema,
                                                                                   transformations=fill_transf)
discretized_frame.head()
```

```python
cols = list(discretized_frame.columns)
```

```python
binarized_frame, bin_schema, transformations = feature_engineering.binarize_all_categorical(discretized_frame,
                                                                                               target='survived',
                                                                                               transformations=ef_transformations)
```

```python
train_cols = list(binarized_frame.columns)
train_cols.remove('survived')
```

```python
X_train = binarized_frame[train_cols].fillna(0).values
Y_train = binarized_frame['survived'].values
```

```python
clf_lr = linear_model.LogisticRegression(C=0.5)
clf_lr.fit(X_train, Y_train)
pmml.log_rgrs_pmml(clf_lr, train_cols, data_dict_list,
                   'survived', 'Log Res for Titanic Survival',
                   'PMML/titanic_LogRes.pmml', transformations=transformations)
```
