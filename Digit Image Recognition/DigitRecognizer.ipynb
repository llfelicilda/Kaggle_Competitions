```python
import numpy as np
import pandas as pd

from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.grid_search import GridSearchCV

from sklearn.decomposition import PCA

from sklearn.preprocessing import MinMaxScaler

from scipy import ndimage

from skimage.filter import threshold_adaptive

from sklearn.cross_validation import KFold

plt.rcParams['figure.figsize'] = (12.0, 12.0)
pd.set_option('max_column', 100)
```

```python
class DataSeparate(object):
    def __init__(self):
        self.LABS = 10             
        
        training_data = pd.read_csv('./data/train.csv', header=0)
        testing_data = pd.read_csv('./data/test.csv', header=0)
        self.xt = np.array(training_data.ix[:, 1:]).astype('float64')
        self.yt = np.atleast_2d(training_data.ix[:, 0]).T

        
        self.xe = self.xt[30000:, :]
        self.ye = self.yt[30000:, :]

        
        self.x_test = np.array(testing_data.ix[:,:]).astype('float64')
```

```python
def process(x_old_set, y_old_set):
    m,n = x_old_set.shape
    
    x_new_set = np.zeros((7*m, n))
    y_new_set = np.zeros((7*m,))
    j = 0
    for i in range(m):
        # add the original one
        x_new_set[j] = x_old_set[i,:]
        y_new_set[j] = y_old_set[i]

        x = x_old_set[i,:].reshape(20,20)
        y = y_old_set[i]

        # move north 1 px
        j +=1
        x_new_set[j] = np.vstack([x[1:,:], x[19:,:]]).flatten()
        #x_new_set[j] = np.roll(x, 2*k[0], k[1]).flatten()
        y_new_set[j] = y
        # move south 1 px
        j +=1
        x_new_set[j] = np.vstack([x[0:1,:], x[0:19,:]]).flatten()
        y_new_set[j] = y
        # move left 1 px
        j +=1
        x_new_set[j] = np.hstack([x[:,1:], x[:,0:1]]).flatten()
        y_new_set[j] = y
        # move right 1 px
        j +=1
        x_new_set[j] = np.hstack([x[:,19:], x[:,0:19]]).flatten()
        y_new_set[j] = y
        # rotate 10 degrees
        j +=1
        x_new_set[j] = ndimage.rotate(x, 10, reshape=False).flatten()
        y_new_set[j] = y
        # rotate -10 degrees
        j +=1
        x_new_set[j] = ndimage.rotate(x, -10, reshape=False).flatten()
        y_new_set[j] = y

        j += 1
    return [x_new_set, y_new_set]
```

```python
def CrossValidateTrain (x_new, y_new, clf):
    kf = KFold(len(x_new), n_folds=10)
    score = 0
    for train_index, test_index in kf:
        X_train, X_test = x_new[train_index], x_new[test_index]
        y_train, y_test = y_new[train_index], y_new[test_index]
        clf.fit(X_train,y_train)
        a = clf.predict(X_test)
        s = clf.score(X_test,y_test)
        if s > score:
            score = s
            predclf = clf
    return predclf
```

```python
def prep_submit():
    clf = SVC(C=6.2, cache_size=200, class_weight=None, coef0=0.0, degree=3, gamma=0.01,
              kernel='rbf', max_iter=-1, probability=False, random_state=None,
                  shrinking=True, tol=0.001, verbose=False)
    clf.fit(x[0:42000,:], y[0:42000])

    x_support = clf.support_vectors_
    y_support = np.array(y[clf.support_])
    [x_new, y_new]= process(x_support, y_support)
    
    predclf = CrossValidateTrain (x_new, y_new, clf)
    a = predclf.predict(x_test)
    b = np.arange(1,28001,1)
    predictions = np.vstack([b, a]).T
    np.savetxt('./submission_svm3.csv', predictions, header='ImageId,Label', fmt='%i,%i')
```

```python
data = DataSeparate()
```

```python
x = data.xt
m = x.shape[0]
new_x = np.zeros((m,400))
for i in range(m):
    old = x[i,:].reshape(28,28)
    block_size = 40
    old = threshold_adaptive(old, block_size, offset=10)
    new_x[i,:] = old[4:24,4:24].flatten()
x = new_x
y = data.yt
x_test = data.x_test
m = x_test.shape[0]
new_x_test = np.zeros((m,400))
for i in range(m):
    old = x_test[i,:].reshape(28,28)
    block_size = 40
    old = threshold_adaptive(old, block_size, offset=10)
    new_x_test[i,:] = old[4:24,4:24].flatten()
x_test = new_x_test
```
