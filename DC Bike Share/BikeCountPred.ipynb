```python
from zeku import feature_engineering, pmml, metrics, data_source, utils, feature_selection
from sklearn import linear_model, tree, ensemble, decomposition, datasets, preprocessing, neighbors,grid_search,cross_validation
from pandas import DataFrame, MultiIndex
import pandas as pd
import numpy as np

pd.set_option('max_column', 100)
```

```python
dcbiketrain_df = data_source.CSVDataSource('DCBikeTrain', 'data/train.csv').data
dcbiketest_df = data_source.CSVDataSource('DCBikeTest', 'data/test.csv').data
```

```python
def processdata (df_):
    df = df_.join(df_[['temp','atemp','humidity','windspeed']].set_index([range(3,len(df_)+3)]), rsuffix = 'adj', how ='left')
    df = df.fillna(method='bfill')
    df[['tempadj','atempadj','humidityadj','windspeedadj']] = df[['tempadj','atempadj','humidityadj','windspeedadj']]**2
    df['Seas_Workd_Weat'] = df['season']*df['workingday']*df['weather']
    df['Seas_Workd'] = df['season']*df['workingday']
    df['Seas_Weat'] = df['season']*df['weather']
    df['Workd_Weat'] = df['workingday']*df['weather']
    df = df.drop(['temp','atemp','humidity','windspeed','season','workingday','weather'],axis=1)
    cols = list(df.columns)
    df0, trans0 = feature_engineering.datetimes.format_datetimes_from_schema(df[cols], datetimes_schema ={'datetime':['%a','%H']})

    cols = list(df0.columns)
    
    df2, df_in_schema, trans2 = feature_engineering.binarize_all_categorical(df0[cols], transformations = trans0)
    return df2
```

```python
def rmsle(pred, actual):
    return np.sqrt(sum(map(lambda x, y: (np.log(x + 1) - np.log(y + 1))**2, pred, actual)) / float(len(actual)))
```

```python
def prediction (pXtrain_df, pYtrain, pXtest_df, model):
    pXtrain_df= preprocessing.normalize(pXtrain_df)
    pXtest_df = preprocessing.normalize(pXtest_df)
    clf1 = model.fit(pXtrain_df,pYtrain)
    kf = cross_validation.KFold(len(pXtrain_df),n_folds=10)
    scores = []
    for train, test in kf:
        X_train, X_test, Y_train, Y_test = pXtrain_df[train],pXtrain_df[test],pYtrain[train],pYtrain[test]
        pred = clf1.predict(X_test).clip(0)
        scores.append(rmsle(Y_test,pred))
    print np.mean(scores)
    return clf1.predict(pXtest_df).clip(0)
```

```python
xtrain = dcbiketrain_df.drop(['casual','count','holiday','registered'],axis=1)
pxtrain = processdata(xtrain)
pxtest = processdata(dcbiketest_df.drop(['holiday'], axis=1))
```

```python
registered = prediction(pxtrain,dcbiketrain_df['registered'].values ,pxtest, ensemble.RandomForestRegressor())
casual = prediction(pxtrain,dcbiketrain_df['casual'].values ,pxtest, ensemble.RandomForestRegressor())
```

```python
registered1 = prediction(pxtrain,dcbiketrain_df['registered'].values ,pxtest, ensemble.GradientBoostingRegressor())
casual1 = prediction(pxtrain,dcbiketrain_df['casual'].values ,pxtest, ensemble.GradientBoostingRegressor())
```

```python
registered2 = prediction(pxtrain,dcbiketrain_df['registered'].values ,pxtest, ensemble.ExtraTreesRegressor())
casual2 = prediction(pxtrain,dcbiketrain_df['casual'].values ,pxtest, ensemble.ExtraTreesRegressor())
```
