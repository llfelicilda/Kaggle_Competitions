```python
from zeku import feature_engineering, pmml, metrics, data_source, utils, feature_selection
from sklearn import linear_model, tree, ensemble, decomposition, datasets, preprocessing, neighbors,grid_search
from pandas import DataFrame, MultiIndex
import pandas as pd
import numpy as np
from ml_metrics import rmsle

pd.set_option('max_column', 100)
```

```python
dcbiketrain_df = data_source.CSVDataSource('DCBikeTrain', 'data/train.csv').data
dcbiketest_df = data_source.CSVDataSource('DCBikeTest', 'data/test.csv').data
```

```python
dcbiketrain_df
```

```python
def processdata (df_):
    df = df_.join(df_[['temp','atemp']][3:], rsuffix = 'adj', how ='left')
    df = df.fillna(method='bfill')
    df = df.drop(['temp','atemp'],axis=1)
    cols = list(df.columns)
    df2, trans2 = feature_engineering.datetimes.format_datetimes_from_schema(df[cols], datetimes_schema ={'datetime':['%a','%H']}
                                                                             )
    cols = list(df2.columns)
    df1, df_in_schema, trans1 = feature_engineering.binarize_all_categorical(df2[cols], transformations = trans2)
    return df1
```

```python
def prediction (pXtrain_df, pYtrain, pXtest_df, model):
    pXtrain_df= preprocessing.normalize(pXtrain_df)
    pXtest_df = preprocessing.normalize(pXtest_df)
    clf = model.fit(pXtrain_df,pYtrain)
    print rmsle(np.around(pYtrain.clip(0)),np.around(clf.predict(pXtrain_df).clip(0)))
    return np.around(clf.predict(pXtest_df).clip(0))
```

```python
xtrain = dcbiketrain_df.drop(['casual','count','registered'],axis=1)
pxtrain = processdata(xtrain)
pxtest = processdata(dcbiketest_df)
```

```python
param_grid= {'n_estimators': range(10,110,10), 'max_depth':range(3,12,2)}
svr = ensemble.RandomForestRegressor()
clf = grid_search.GridSearchCV(svr,param_grid,cv=10, verbose=2).fit(pxtrain,dcbiketrain_df['registered'].values)
```

```python
print 'Best parameters set found on the development set:'
print clf.best_estimator_
```

```python
param_grid= {'n_estimators': range(10,110,10), 'max_depth':range(3,12,2)}
svr = ensemble.ExtraTreesRegressor()
clf = grid_search.GridSearchCV(svr,param_grid,cv=10, verbose=2).fit(pxtrain,dcbiketrain_df['casual'].values)
print 'Best parameters set found on the development set:'
print clf.best_estimator_
```

```python
registered = prediction(pxtrain,dcbiketrain_df['registered'].values ,pxtest, ensemble.ExtraTreesRegressor())
casual = prediction(pxtrain,dcbiketrain_df['casual'].values ,pxtest, ensemble.ExtraTreesRegressor())
```

```python
registered = prediction(pxtrain,dcbiketrain_df['registered'].values ,pxtest, ensemble.RandomForestRegressor())
casual = prediction(pxtrain,dcbiketrain_df['casual'].values ,pxtest, ensemble.RandomForestRegressor())
```

```python
pd.DataFrame({'count':registered+casual, 'datetime':dcbiketest_df.datetime}, columns=['datetime','count']).to_csv('bike.csv', index=False)
```
